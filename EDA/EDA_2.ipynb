```python
# import libtries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [10, 7]
plt.rcParams['font.size'] = 14
plt.rcParams['font.weight'] = 'bold'
plt.style.use('seaborn-white')
```

```python
# Read in the data set
insurance = pd.read_csv('C:/Users/tohaj/Box/Programming/Data Science/DATA SETS/insurance.csv')
```

```python
# View the 1st 7 observations
insurance.head(7)
```

```python
# View the last 5 observations 
insurance.tail()
```

```python
# View the shape of (i.e., no. of rows and cols in the data)

# Method 1: using '+' and the str() function
print('Number of rows and columns in the data: ' + str(insurance.shape))

# Method 2: using ','
print('\nNumber of rows and columns in the data: ', insurance.shape)
# NB: the leading '\n' created an empty line before the printed word
```

```python
# Create a linear plot (regression) of Expenses vs. BMI
sns.lmplot(x = 'bmi', y = 'expenses', data = insurance, aspect = 1.25, height = 8)
plt.title('Linear plot of Expenses vs. BMI')
plt.xlabel('Body mass index (in $kg/m^2$)')
plt.ylabel('Insurance expenses (in $USD$)')
```

```python
# View the basic stats of the data
insurance.describe()
#  this function applies to only the quantitative columns
```

```python
# Provide a brief info of the data
insurance.info()
# this function applies to all columns/variables in the data
```

```python
# Check for missing values in the data
insurance.isnull()
```

```python
# Check for missing values in the 1st 10 obs
insurance.head(10).isnull()
```

```python
# Plot a heatmap of all mssing values in the data
plt.figure(figsize = (10, 7))
sns.heatmap(insurance.isnull(), cmap = 'rainbow', yticklabels = False)
# NB: the color bar to the right is a legend that helps read values on the heatmap. There is no missing values at all, so the entire map is green
#...(the color value for 0 as shown in the color bar). This parameter can be switched off by including 'cbar = False'
plt.title('Checking for missing values in the data')
```

```python
# Create a heatmap displaying the pairwise correlation of the entire data set
# NB: Because it is correlation, only the quant variables are included
plt.figure(figsize = (10, 6))
sns.heatmap(insurance.corr(), cmap = 'rainbow', annot = True)
plt.title('Heatmap showing pairwise correlation of the qunatitative variables')
```

```python
# Plot univariate distribution graphs of expenses, BMI, and age

t = plt.figure(figsize = (12, 10))

# Expenses
ax_1 = t.add_subplot(321)
sns.distplot(insurance['expenses'], bins = 40, color = 'black', ax = ax_1)
ax_1.set_title('Univariate distribution of Expenses')

ax_2 = t.add_subplot(322)
sns.distplot(np.log(insurance['expenses']), bins = 40, color = 'purple', ax = ax_2)
ax_2.set_title('Univariate distribution of Expenses in $log$ scale')

# BMI
ax_3 = t.add_subplot(323)
sns.distplot(insurance['bmi'], bins = 40, color = 'red', ax = ax_3)
ax_3.set_title('Univariate distribution of BMI')

ax_4 = t.add_subplot(324)
sns.distplot(np.log(insurance['bmi']), bins = 40, color = 'blue', ax = ax_4)
ax_4.set_title('Univariate distribution of BMI in $log$ scale')

# Age
ax_5 = t.add_subplot(325)
sns.distplot(insurance['age'], bins = 40, color = 'yellow', ax = ax_5)
ax_5.set_title('Univariate distribution of Age')

ax_6 = t.add_subplot(326)
sns.distplot(np.log(insurance['age']), bins = 40, color = 'brown', ax = ax_6)
ax_6.set_title('Univariate distribution of Age in $log$ scale')

t.tight_layout()
```

```python
# Create a violin plots of of expenses by sex and by smoker, both on the same plotting area
b = plt.figure(figsize = (12, 6))

ax_i = b.add_subplot(121)
sns.violinplot(x = 'sex', y = 'expenses', data = insurance, palette = 'viridis', ax = ax_i)
ax_i.set_title('Violin plot of expenses by sex')

ax_ii = b.add_subplot(122)
sns.violinplot(x = 'smoker', y = 'expenses', data = insurance, palette = 'rainbow', ax = ax_ii)
ax_ii.set_title('Violin plot of expenses by smoker')

b.tight_layout()
```

```python
# Create a violin plot of Expenses vs. Children, grouped by Sex
sns.violinplot(x = 'children', y = 'expenses', data = insurance, hue = 'sex', palette = 'turbo')
plt.title('Violin plot of Expenses vs. Children, grouped by Sex')
```

```python

# Create a violin plot of Expenses vs. Sex, grouped by Children
sns.violinplot(x = 'sex', y = 'expenses', data = insurance, hue = 'children', palette = 'turbo_r')
plt.title('Violin plot of Expenses vs. Sex, grouped by Children')
```

```python
# Insert a cut point in the latest violin plot, so that there are no negative values
# Create a violin plot of Expenses vs. Sex, grouped by Children
sns.violinplot(x = 'sex', y = 'expenses', data = insurance, hue = 'children', palette = 'turbo_r', cut = 0)
plt.title('Violin plot of Expenses vs. Sex, grouped by Children: \n Cut-off at zero')
```

```python
# Create a box plot of Expenses vs. Children, grouped by Sex
plt.figure(figsize = (12, 9))
sns.boxplot(x = 'children', y = 'expenses', data = insurance, hue = 'sex', palette = 'inferno')
plt.title('Box plot of Expenses vs. Children, grouped by Sex')
```

```python
# Create violin plot of of Expenses vs. Region, grouped by sex (categories combined)
plt.figure(figsize = (10, 8))
sns.violinplot(y = 'expenses', x = 'region', data = insurance, hue = 'sex', split = True)
plt.title('Violin plot of of Expenses vs. Region grouped by sex: \n Sex categories combined onto a single violin')
```

```python
v = plt.figure(figsize = (13, 7))

# Plot a scatter of Expenses vs. Age
ax_01 = v.add_subplot(121)
sns.scatterplot(x = 'age', y = 'expenses', data = insurance, palette = 'twilight', ax = ax_01)
ax_01.set_title('Scatter plot of Expenses vs. Age')

# Plot a scatter of Expenses vs. BMI
ax_02 = v.add_subplot(122)
sns.scatterplot('bmi', 'expenses', data = insurance, palette = 'twilight_r', ax = ax_02)
ax_02.set_title('Scatter plot of Expenses vs. BMI')
```

```python
# Dummy-code all categorical variables in the data
insurance_recoded = pd.get_dummies(data = insurance, prefix = 'OHE', prefix_sep = '-', columns = ['sex', 'children', 'smoker', 'region'],
    drop_first = True, dtype = 'int8')
```

```python
# Print out the columns and the dimensions of both dataset
print('The shape/dimension of the original data is: ' + str(insurance.shape))
print('\nThe following are the column names in the original data: \n', insurance.columns.values)
print('\nThe shape/dimension of the new data is: ', insurance_recoded.shape)
print('\nThe following are the column names in the new data: \n'+ str(insurance_recoded.columns.values))
```

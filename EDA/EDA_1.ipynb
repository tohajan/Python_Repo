```python
# create a linear plot (regression graph) of expenses against BMI
sns.lmplot(x = 'bmi', y = 'expenses', data = df, aspect = 2, height = 5)
plt.xlabel('BMI (in $kg/m^2$): as independent variable')
plt.ylabel("Expenses (in $USD): as dependent variable")
plt.title("Graph of Expenses vs. BMI")
```

```python
# plot a heatmap of missing values
plt.figure(figsize= (8, 6))

sns.heatmap(df.isnull(), cmap = 'viridis', cbar = False, yticklabels = False)

plt.title('Checking for missing values')
```

```python
# Running a pairwise correlation of all quantitative variables

plt.figure(figsize = (12, 8))

sns.heatmap(df.corr(), cmap = "Paired_r", annot = True, cbar = False)

plt.title("Pairwise correlation of all quantitative variables")
```

```python
# Plot univariate distributions (raw and log) of Expenses
g = plt.figure(figsize = (14, 8))

ax1 = g.add_subplot(221)
sns.distplot(df['expenses'], bins = 50, color = "r", ax = ax1)
ax1.set_title('Distribution of Expenses')

ax2 = g.add_subplot(222)
sns.distplot(np.log(df['expenses']), bins = 50, color = "b", ax = ax2)
ax2.set_title('Distribution of Expenses in $log$ scale')


# Plot univariate distributions (raw and log) of BMI
ax3 = g.add_subplot(223)
sns.distplot(df['bmi'], bins = 50, color = "g", ax = ax3)
ax3.set_title('Distribution of BMI')

ax4 = g.add_subplot(224)
sns.distplot(np.log(df['bmi']), bins = 50, color = "k", ax = ax4)
ax4.set_title('Distribution of BMI in $log$ scale')

g.tight_layout() 
# this spaces out the plots so that the axes (or axis labels) don't overlap
# Note that the code is placed after all participating plots have been added. It doesn't work otherwise
```

```python
h = plt.figure(figsize = (20, 10))

# Create a violin plot of expenses vs. sex
ax_a = h.add_subplot(121)
sns.violinplot(x = "sex", y = "expenses", data = df, palette = "magma", ax = ax_a)
ax_a.set_title("Violin plot of Expenses vs. Sex")

# Add a violin plot of expenses vs. smoking status
ax_b = h.add_subplot(122)
sns.violinplot(x = "smoker", y = "expenses", data = df, palette = "Wistia", ax = ax_b)
ax_b.set_title("Violin plot of Expenses vs. Smoking")
```

```python
# Create a violin plot of expenses vs. children, grouped by sex

plt.figure(figsize = (15, 10))
sns.violinplot(x = "children", y = "expenses", data = df, hue = "sex", palette = "turbo")
plt.title("Violin plot of Expenses vs. Children, grouped by sex")
```

```python
# Because violin plots are estimate of actual values, there is almost a;ways "negative" values (these are estimates of actual values that
# ...are very close to 0).
# Add a cutpoint (at 0) to cut-off the seemingly negative "readings"

plt.figure(figsize = (13, 9))
sns.violinplot(x = "children", y = "expenses", data = df, hue = "sex", palette = "turbo", cut = 0)
plt.title("Violin plot of Expenses vs. Children, grouped by sex")
```

```python
# Draw a box plot of expenses vs. children, grouped by sex
plt.figure(figsize = (13, 10))
sns.boxplot(x = "children", y = "expenses", data = df, palette = "rainbow", hue = "sex")
plt.title("Box plot of Expenses vs. Children, grouped by Sex")
```

```python
# Create a violin plot of expenses vs. region, grouped by sex (with categories combined on a single violin)
plt.figure(figsize = (13, 10))
sns.violinplot(x = "region", y = "expenses", data = df, hue = "sex", palette = "twilight", split = True)
plt.title("violin plot of Expenses vs. Region: \nSex categories combined on a single violin")
```

```python
q = plt.figure(figsize = (15, 10))

# Create a scatterplot of expenses vs. age
ax_i = q.add_subplot(121)
sns.scatterplot(x = "age", y = "expenses", data = df, ax = ax_i, hue = "smoker", palette = "twilight_shifted")
ax_i.set_title("Scatter plot of Expenses vs. Age")

# Add a scatterplot of expenses vs. bmi
ax_ii = q.add_subplot(122)
sns.scatterplot(x = "bmi", y = "expenses", data = df, ax = ax_ii, hue = "smoker", palette = "twilight_r")
ax_ii.set_title("Scatter plot of Expenses vs. BMI")
```

```python
# Dummy-recode all categorical variables in the data set

categorical_columns = ['sex', 'children', 'smoker', 'region']
df_recode = pd.get_dummies(data = df, prefix = 'dummy', prefix_sep = '_', columns = categorical_columns, drop_first = True, dtype = 'int8')
```
